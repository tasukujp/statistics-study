```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.title("scatterplot")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
x = np.random.normal(size = 2000)
y = np.random.normal(size = 2000)

plt.scatter(x, y, alpha = 0.3)

plt.title("scatterplot")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

farm = pd.read_csv('farm.csv')
plt.scatter(farm['sales'],farm['area']) #  http://stat.biopapyrus.net/python/scatter.html
```

```python
import numpy as np

print("")
print(" :" + str(np.average(farm['sales']))) #  http://tokeigaku.blog.jp/python/numpy/%E5%B9%B3%E5%9D%87
print(" :" + str(np.var(farm['sales']))) #  http://tokeigaku.blog.jp/python/numpy/%E5%88%86%E6%95%A3
print(" :" + str(np.std(farm['sales']))) #  http://tokeigaku.blog.jp/python/numpy/%E6%A8%99%E6%BA%96%E5%81%8F%E5%B7%AE
print(" :" + str(np.std(farm['sales'])/np.average(farm['sales']))) # 

print("")
print(" :" + str(np.average(farm['area'])))
print(" :" + str(np.var(farm['area'])))
print(" :" + str(np.std(farm['area'])))
print(" :" + str(np.std(farm['area'])/np.average(farm['area'])))
```
